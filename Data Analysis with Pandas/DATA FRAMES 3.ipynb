# FILTERING WITH THE .WHERE() METHOD

```python
bond=pd.read_csv("jamesbond.csv",index_col="Film")
bond.sort_index(inplace=True)
bond.head()
```

```python
actor = bond["Actor"] == "Sean Connery"
bond[actor]
```

```python
bond.where(actor).head()
```

```python
box_office = bond["Box Office"] > 800
```

```python
bond.where(box_office).head(10)
```

```python
bond.where(actor & box_office)
```

# WHERE METHOD IS VERY EFFECTIVE - TWO CONDITIONS WHERE ACTOR = SEAN CONNERY AND BOX OFFICE > 800 CAN BE SEEN ABOVE

# THE .QUERY() METHOD

```python
bond=pd.read_csv("jamesbond.csv",index_col="Film")
bond.sort_index(inplace=True)
bond.head()
```

# TWO THINGS TO REMEMBER

# if there are spaces between columns it does not work 

# argument will only be a string

```python
bond.columns = [column_name.replace(" ","_") for column_name in bond.columns]
```

```python
bond.columns
```

```python
bond.head()
```

```python
bond.query('Actor == "Sean Connery"')
```

```python
bond.query('Year == "1967"')
```

```python
bond.query("Actor == 'Roger Moore' and Director == 'John Glen'")
```

```python
bond.query("Actor == 'Roger Moore' or Director == 'John Glen'")
```

```python
bond.query("Actor in ['Roger Moore','Timothy Dalton']")
```

```python
bond.query("Box_Office > 600")
```

```python
bond.query("Actor in ['Roger Moore','Daniel Craig']"  or "Director == John Glen")
```

```python
bond.query("Actor not in ['Roger Moore','Daniel Craig','Sean Connery']")
```
