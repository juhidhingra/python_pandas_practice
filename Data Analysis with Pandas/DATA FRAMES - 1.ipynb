```python
complete_null_row = basket_ball_team.dropna(how='all')
```

```python
basket_ball_team.tail(3)
```

# DROP COLUMNS WITH NULL VALUES

```python
salary.head()
```

# By specifying dtype="category" when constructing a Series:

#In [1]: s = pd.Series(["a","b","c","a"], dtype="category")

#In [2]: s
#Out[2]: 
#0    a
#1    b
#2    c
#3    a
#dtype: category
#Categories (3, object): [a, b, c]

```python
basket_ball_team["Position"].nunique()
```

```python
basket_ball_team["Position"] = basket_ball_team["Position"].astype("category")
```

```python
basket_ball_team.info()
```

# reduces memory from 35.7 kb to 27.4 kb => makes sense when using large data sets it helps reduce memory usage

```python
basket_ball_team["Team"] = basket_ball_team["Team"].astype("category")
```

```python
basket_ball_team.info()
```

```python
# memory usage further reduced 
```

# RANK VALUES WITH THE .RANK() METHOD

```python
basket_ball_team = pd.read_csv("nba.csv")
```

```python
basket_ball_team.head()
```

```python
columns_with_complete_null_values = basket_ball_team.dropna(how='all')
```

```python
columns_with_complete_null_values.head()
```

# dropna - how = 'all' removes column with ALL NULL VALUES

```python
basket_ball_team["Salary"] = basket_ball_team["Salary"].fillna(value=0).astype("int")
```

```python
basket_ball_team.head(3)
```

```python
basket_ball_team["Salary Rank"] = basket_ball_team["Salary"].rank(ascending=False).astype("int")
```

```python
basket_ball_team.head()
```

# ascending = False -> highest salary are going to possess the smaller ranks 

```python
sorting_salary_column = basket_ball_team.sort_values(by = "Salary",ascending=False)
```

```python
sorting_salary_column.head()
```

# highest salary has rank 1
