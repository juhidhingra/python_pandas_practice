# SORT_INDEX METHOD called on multi index dataframe

```python
mac = pd.read_csv("bigmac.csv",parse_dates=["Date"],index_col=["Date","Country"])
mac.sort_index(inplace=True)
mac.head(3)
```

```python
mac.sort_index(ascending=[True,False],inplace=True)
mac.head(3)
```

```python
mac.sort_index(ascending=[False,True],inplace=True)
mac.head(3)
```

# IN THE ABOVE METHOD YOU CAN SET THE ORDER OF ASCENDING VALUES - SINCE THERE ARE TWO LAYERS YOU CAN SET THE ORDER IN A LIST --- DATE = FALSE, COUNTRY = TRUE

# EXTRACT ROWS FROM MULTIINDEX DATA FRAME

```python
mac = pd.read_csv("bigmac.csv",parse_dates=["Date"],index_col=["Date","Country"])
mac.sort_index(inplace=True)
mac.head(3)
```

# the loc IS ONE OF THE RARE METHODS THAT TAKE SQUARE BRACKETS AND DUE TO TECHNICAL DESIGN LOC DOES NOT TAKE A LIST BUT YOU CAN PASS A TUPLE IN IT.. 

# A TUPLE IS VERY SIMILAR TO A LIST - ONLY DIFFERENCE - ITS IMMUTABLE - CANNOT BE MODIFIED AFTER CREATION

# argument 1 in brackets is on horizontal axis and 2 argument after the comma is on vertical axis

```python
mac.loc[("2010-01-01","Argentina"),"Price in US Dollars"]
```

# HERE WE ARE EXTRACTING ROW FROM MUTLI INDEX

```python
mac.loc[("2015-07-01")]
```

```python
mac.loc[("2015-07-01","Chile"),"Price in US Dollars"]
```

# above you get a series for country chile

# THE TRANSPOSE METHOD

```python
mac = pd.read_csv("bigmac.csv",parse_dates=["Date"],index_col=["Date","Country"])
mac.sort_index(inplace=True)
mac.head(3)
```

# TRANSPOSE BASICALLY REVERSES YOUR AXIS 

```python
mac.transpose()
```

# transpose doesnt have the inplace parameter so you can reassign the values to mac.transpose by calling mac = mac.transpose()

```python
mac = mac.transpose()
```

```python
mac.head(3)
```

# THE SWAPLEVEL METHOD()

```python
mac = pd.read_csv("bigmac.csv",parse_dates=["Date"],index_col=["Date","Country"])
mac.sort_index(inplace=True)
mac.head(3)
```

# TO SWAP YOUR INDEX POSITION
