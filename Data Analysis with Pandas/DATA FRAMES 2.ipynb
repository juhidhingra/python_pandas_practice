# THE .DUPLICATED() METHOD

```python
df = pd.read_csv("employees.csv", parse_dates=["Start Date","Last Login Time"])

df["Senior Management"]= df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.sort_values("First Name",inplace=True)
df.head()
```

# all items after the first one are marked as duplicate

```python
df["First Name"].duplicated().head()
```

```python
df["First Name"].duplicated(keep="last").head()
```

# the above method keep = true keeps all values that are the last values in duplicate list and remove the ones before that

# now to remove all duplicate values

# keep = false will mark eveything as dupliacted if it occurs more than once no matter if its the first value or the last value

```python
df["First Name"].duplicated(keep=False).head()
```

# to FIND ONLY UNIQUE VALUES - USE THE TILDE SYMBOL

```python
unique_values = -df["First Name"].duplicated(keep=False)
```

```python
df[unique_values]
```

# these are all the names that occur once and only once in the data set

# DROP DUPLICATES() METHOD

```python
df = pd.read_csv("employees.csv", parse_dates=["Start Date","Last Login Time"])

df["Senior Management"]= df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.sort_values("First Name",inplace=True)
df.head()
```

# two rows were entirely identitical for drop duplicates to show duplicates,tehrefore no rows are being removed by this method below

```python
len(df)
```

```python
len(df.drop_duplicates())
```

# subset accepts a list of strings & those strings can represent names of the columns & keep = 'first' means it keeps the 1st occurence of a value even if it is a duplicate value

```python
df.drop_duplicates(subset=["First Name"],keep='first').head()
```

# keep= first - keeps the first duplicate value and removes other
# keep = last - keeps the last duplicate value and remove other

# if we want to remove row if there is duplicate, JUST MAKE KEEP = FALSE

```python
df.drop_duplicates(subset=["First Name"],keep=False)
```

# if there is no unique value in a data it will just return the column names without any values

```python
df.drop_duplicates(subset=["Team"],keep=False)
```

# there was no team that was unique it had all dupliactes

# use on two columns will make the drop_duplicates to look at both the column before dropping any duplicate values

```python
df.drop_duplicates(subset=["First Name","Team"]).head()
```

# aaron is twice but their team is different ie.unique for both, so it does not remove that value
