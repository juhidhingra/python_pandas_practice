# ITERATING THROUGH GROUPS

```python
fortune =pd.read_csv("fortune1000.csv",index_col="Rank")
sectors = fortune.groupby(by = ["Sector","Industry"])
fortune.head()
```

# want to extract highest profit or highest revenue 

```python
sectors["Profits"].max()
```

# we need to iterate as we need multiple things done 

```python
df = pd.DataFrame(columns=fortune.columns)
```

```python
df
```

# next step - looping through all our groups and getting highest revenue

# you always have to provide two variable names each time you iterate over a group by object

# 1st variable - to represent the group
# 2nd variable - to represent the actual dataframe that is STORED WITHIN THAT GROUPING

# nlargest- we want to specify the number of values we want, in this case only the top value, then the second argument you specify which column

# data is temporary VARIABLE name -for each one of our 21 data frames 

# append will just add everything together

```python
for sector, data in sectors:
    
    highest_revenue_company_in_group = data.nlargest(1,"Revenue")
    df = df.append(highest_revenue_company_in_group)
```

```python
df
```

# the above dataframe shows the highest revenue making 21 companies
