# PIVOT TABLE METHOD()

# works the same way as pivot in excel

```python
food =pd.read_csv("foods.csv")
```

```python
food.head()
```

# always a good idea to use numeric columns in values 

```python
food.pivot_table(values="Spend",index="Gender",aggfunc="sum")
```

```python
food["Gender"].value_counts()
```

```python
food.pivot_table(values="Spend",index="Gender",aggfunc="mean")
```

```python
food.pivot_table(values="Spend",index=["Gender","Item"],columns=["City"],aggfunc="sum")
```

```python
food.pivot_table(values=["Spend"],index=["Gender","Item"],columns= ["City"],aggfunc="count")
```

```python
food.pivot_table(values=["Spend"],index=["Item","Gender"],columns= ["City"],aggfunc=["max"]).head()
```

# YOU CAN CALL THE PIVOT TABLE METHOD DIRECTLY ON PANDAS

```python
pd.pivot_table(data=food,values=["Spend"],index=["Item","Gender"],columns= ["City"],aggfunc=["max"]).head()
```

# THE PD.MELT() METHOD

# WORKS AS A REVERSE OPERATION TO A PIVOT TABLE AND TAKES AN AGGREGATED DATA SET & CONVERST IT TO A TABULAR FORMAT

```python
sales= pd.read_csv("quarters.csv")
```

```python
sales
```

# melts 4 columns into one row 

```python
pd.melt(frame= sales, id_vars= "Salesman")
```

```python
pd.melt(frame=sales, id_vars="Salesman",var_name="Quarter",value_name="Revenue")
```
